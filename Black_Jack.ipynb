```python
# Initialization

import random
from IPython.display import clear_output

# Classes
 
class CARD:
    def __init__(self,suit,rank):
        # internal parameters
        dic_ranks = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Prince':10, 'Qeen':10, 'King':10, 'Ace':11}
        # attributes
        self.suit = suit
        self.rank = rank
        self.value = dic_ranks[rank]
        
    def __str__(self):
        return f"{self.rank} of {self.suit}"
```

```python
class DECK:   
    # Class object attributes
    pass

    def __init__(self):
        # internal parameters
        suits = ['Hearts','Diamonds','Spades','Clubs']
        ranks = ['Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Prince', 'Qeen', 'King', 'Ace']
        # attributes
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                self.all_cards.append(CARD(suit,rank))
        print(f"A deck with {self.size()} cards was created\n")
        
    def __str__(self):
        return f"The deck has {self.size()} cards\n"

    # methods   
    def size(self):
        return len(self.all_cards)
    
    def shuffle(self):
        random.shuffle(self.all_cards)
        print("shuffling...\nshuffling...\nshuffling...\n\nThe deck is shuffled\n")

    def deal(self,serie_length=1):
        if self.size() == 0:
            print("No more cards in the deck\n")
            return -1
        else:
            cards = []
            for num in range(serie_length):
                 cards.append(self.all_cards.pop(0))
            return cards
    
    def print_all_cards(self):
        for card in range(self.size()):
            print(self.all_cards[card])
```

```python
class PLAYER:    
    def __init__(self):
        # attributes
        self.name = input("Hi there player, what's your name? ")
        self.account = 1000000
        self.hand = []
        self.bet_amount = 0
        print(f"\nWelcome {self.name} !\nYou have {self.account/1000000:1.0f} million $ for betting in this game.\nEnjoy :)\n")
    
    # methods
    def __str__(self):
        return f"Player's name is {self.name}\nPlayer's balance is {self.account} $\n"
        
    def bet(self):
        while True:
            try:
                bet = int(input(f"Your cash balance is {self.account} $\nhow much would you like to bet in your next hand? "))
                print('\n')
            except:
                print("oops...\nUse numbers in your answer")
                continue
            else:
                if bet > self.account:
                    print("hmmm...\nYou can't bet more than you have got")
                    continue
                elif bet == 0:
                    print("Come on... You need to bet if you wanna play!")
                    continue
                else:
                    break
        self.account -= bet
        self.bet_amount = bet
        
    def get_money(self,money):
        self.account += money
        print(f"{self.name}, you got {money} $")
    
    def stand(self):
        clear_output()
        print('Player stands\n')
    
    def hit(self,DECK,number_of_cards=1):
        self.hand.extend(DECK.deal(number_of_cards))
    
    def check_hand(self):
        total_value = 0
        ace_count = 0
        for CARD in self.hand:
            total_value += CARD.value
            if CARD.value == 11: ace_count += 1
        for index in range(ace_count):
            if total_value > 21:
                total_value -= 10
        return total_value
    
    def display_hand(self):
        print(f"{self.name}, Your hand has:")
        for CARD in range(len(self.hand)):
            print(self.hand[CARD])
        print("\n")
```
